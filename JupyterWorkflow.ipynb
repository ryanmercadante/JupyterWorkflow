#### From exploratory analysis to reproducible science

```python
URL = 'https://data.seattle.gov/api/views/65db-xm6k/rows.csv?accessType=DOWNLOAD'
```

```python
from urllib.request import urlretrieve
urlretrieve(URL, 'Fremont.csv')
```

```python
import pandas as pd
data = pd.read_csv('Fremont.csv', index_col='Date', parse_dates=True)
data.head()
```

```python
%matplotlib inline
# Get the total number of rides each week over the course of each year
data.resample('W').sum().plot()
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
data.columns = ['West', 'East']
data.resample('W').sum().plot()
```

```python
# This time we're showing the sum of rides in the previous 365 days
data.resample('D').sum().rolling(365).sum().plot()
```

```python
data['Total'] = data['West'] + data['East']

# Reset the y-axis so that is scales to 0
ax = data.resample('D').sum().rolling(365).sum().plot()
ax.set_ylim(0, None) # Using None as the second parameter says to use the current maximum
```

```python
data.groupby(data.index.time).mean().plot()
```

```python
pivoted = data.pivot_table('Total', index=data.index.time, columns=data.index.date)
pivoted.iloc[:5, :5]
```

```python
pivoted.plot(legend=False, alpha=0.01)
```
